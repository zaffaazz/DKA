```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Support function for printing graphs
def show_graph(G, pos=None, title='') :
  # Create a post if a post is not given
  if pos is None:
      pos = nx.spring_layout(G)

  # Function to draw nodes
  nx.draw(
      G,                  # Graf NetworkX
      pos,                # Node position
      with_labels=True,   # Displays the code name
      node_color='pink',   # Node color
      node_size=2000,     # Node size
      font_color="white", # Node label font color
      font_weight="bold", # Node label font thickness
      width=5             # Edge line thickness
  )

  # Takes edge labels if there is weight
  edge_labels = nx.get_edge_attributes(G, 'weight')
  # Function to draw nodes
  nx.draw_networkx_edge_labels(
      G,
      pos,
      edge_labels=edge_labels, # Data weight
      font_color='blue',       # Edge label font color
      font_weight="bold",      # Edge label font weight
      font_size=12,            # Edge label font size
  )

  plt.margins(0.2)  # Provides margin to the plot
  plt.title(title)  # Displays the graph title if given
  plt.show()
```

```python
# Initialize empty graph
G = nx.Graph()
```

```python
edges = [
    ('S', 'A', 10),
    ('S', 'B', 20),
    ('S', 'C', 15),
    ('A', 'D', 25),
    ('B', 'D', 5),
    ('B', 'E', 10),
    ('C', 'E', 25),
    ('D', 'G', 20),
    ('E', 'G', 5),
]
G.add_weighted_edges_from(edges)
```

```python
show_graph(G, pos=pos)
```

```python
print("Sequence of visited nodes:")  # DFS

i = 1
parent = {'S': None}

for nodes in nx.dfs_edges(G, source='S'):
    print(f"{i}. {nodes[0]} - {nodes[1]}")
    if nodes[1] not in parent:
        parent[nodes[1]] = nodes[0]
    if nodes[1] == 'G':
        break
    i += 1
path = []
node = 'G'
while node is not None:
    path.append(node)
    node = parent[node]
path.reverse()
cost = 0
for i in range(len(path) - 1):
    cost += G[path[i]][path[i+1]]['weight']

print("Solusi Jalur:", path)
print("Total Cost   :", cost)
```

```python
def distance(node1, node2):
    for neighbor, cost in graph[node1]:
        if neighbor == node2:
            return cost
    return float('inf')

# Hitung total cost sebuah path
def total_distance(path, graph):
    dist = 0
    for i in range(len(path) - 1):
        dist += distance(path[i], path[i+1])
    return dist

# Swap 2 node di tengah path
def random_swap(path):
    new_path = path[:]
    if len(new_path) > 3:
        i, j = random.sample(range(1, len(path) - 1), 2)
        new_path[i], new_path[j] = new_path[j], new_path[i]
    return new_path
```

```python
def simulated_annealing(graph, start, goal, initial_temp, cooling_rate, stopping_temp):
    def greedy_path(start, goal):
        path = [start]
        visited = {start}
        current = start
        while current != goal:
            neighbors = [(n, heuristic[n]) for n, _ in graph[current] if n not in visited]
            if not neighbors:
                break
            current = min(neighbors, key=lambda x: x[1])[0]
            path.append(current)
            visited.add(current)
        return path

    current_solution = greedy_path(start, goal)
    current_distance = total_distance(current_solution, graph)
    best_solution = current_solution[:]
    best_distance = current_distance
    temp = initial_temp

    print("=" * 50)
    print("SIMULATED ANNEALING")
    print(f"Start: {start} | Goal: {goal}")
    print(f"Solusi Awal : {' -> '.join(current_solution)}")
    print(f"Cost Awal   : {current_distance}")
    print("=" * 50)

    while temp > stopping_temp:
        new_solution = random_swap(current_solution)
        valid = all(distance(new_solution[i], new_solution[i+1]) != float('inf')
                    for i in range(len(new_solution)-1))
        if not valid:
            temp *= cooling_rate
            continue

        new_distance = total_distance(new_solution, graph)

        if new_distance < current_distance:
            current_solution = new_solution[:]
            current_distance = new_distance
        else:
            if random.random() < math.exp((current_distance - new_distance) / temp):
                current_solution = new_solution[:]
                current_distance = new_distance

        if current_distance < best_distance:
            best_solution = current_solution[:]
            best_distance = current_distance

        temp *= cooling_rate

    print("\nHASIL AKHIR:")
    print(f"Jalur Solusi : {' -> '.join(best_solution)}")
    print(f"Actual Cost  : {best_distance}")
    return best_solution, best_distance
```
