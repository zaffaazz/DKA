```python
import matplotlib.pyplot as plt
# Data
fig = plt.figure(figsize=(8,6))

#menambah plot
ax = fig.add_subplot(1,1,1)
#membuat plot
ax.plot([1,2,3,4],[10,20,25,30],label='Data1')

#menambahkan nama ke tabel
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Contoh plot dan Figure")

ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
kategori=['A','B','C','D']
nilai = [10,20,25,30]

#membuat figure
fig=plt.figure(figsize=(6,8)) #lebar 10 inci dan tinggi 6 inci


plt.bar(kategori, nilai, hatch='/', color='pink')

#menambahkan nama ke tabel
plt.xlabel("Data X")
plt.ylabel("Data Y")
plt.title("Contoh bar plot")


#menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
x=[1,2,3,4]
y1=[9,20,27,35]
y_main=[7,18,22,29]
y2=[6,15,18,23]

#membuat plot
plt.figure(figsize=(10,6)) #lebar 10 inci dan tinggi 6 inci

#memebuat fill_between plot
plt.fill_between(x, y1,y2, color='pink', alpha=0.5, label='Fill Plot')

plt.plot(x, y_main, color='red', linewidth=2, label='Garis Utama')

#menambahkan nama ke tabel
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Contoh Fill between")

#
plt.legend()

#menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
x=[1,2,3,4]
y1=[10,20,25,30]
y2=[5,10,15,20]
y3=[2,5,8,12]

#membuat plot
plt.figure(figsize=(10,6)) #lebar 10 inci dan tinggi 6 inci

#memebuat fill_between plot
plt.stackplot(x, y1, y2, y3, labels=['Y1', 'Y2','Y3'])
#menambahkan nama ke tabel
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Stack Plot")

#
plt.legend()

#menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [7,5,8,9,6,7,5,8,6,7,6,8,9,4,7,5,9,8,6,7]


#membuat plot
fig=plt.figure(figsize=(8,6)) #lebar 10 inci dan tinggi 6 inci

plt.boxplot(data, patch_artist=True, widths=0.50) #lebar boc = 0.50
#menambahkan nama ke tabel

plt.ylabel("Nilai")
plt.title("Box Plot")

#menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [7,5,8,9,6,7,5,8,6,7,6,8,9,4,7,5,9,8,6,7]


#membuat plot
fig=plt.figure(figsize=(6,8)) #lebar 10 inci dan tinggi 6 inci

plt.violinplot(data, showmeans=True, showmedians=True) #lebar boc = 0.5
#menambahkan nama ke tabel

plt.ylabel("Nilai")
plt.title("Violin Plot")

#menampilkan
plt.show()
```

```python
import matplotlib.pyplot as plt

label = ['A', 'B', 'C', 'D', 'E']
nilai = [15,25,35,20,5]


#membuat plot
plt.figure(figsize=(8,8)) #lebar 10 inci dan tinggi 6 inci

plt.pie(nilai, labels=label, autopct='%1.2f%%') #lebar boc = 0.5
#menambahkan nama ke tabel

plt.title("Contoh Pie Chart")

plt.legend()
#menampilkan
plt.show()
```
